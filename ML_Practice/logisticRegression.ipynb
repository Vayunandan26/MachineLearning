```python
def preprocessed (df):
    #we need to create functions for label encoding internship experience and 
    # removing cgpa values above 10 which could be spelling mistakes
    #also use target encoding for the college ID class 
    df['CGPA'] = df.loc[df['CGPA'] > 10, 'CGPA'] = np.nan
    modified_mean = df['CGPA'].mean
    df.fillna({'CGPA':modified_mean}, inplace= True)
    
```

```python
shuffle_df = df.sample(frac=1, random_state=42)
train_ratio = 0.80
split_index = int(train_ratio*len(shuffle_df))
train_df = shuffle_df[:split_index].reset_index(drop=True)
test_df = shuffle_df[split_index:].reset_index(drop=True)
```

```python
train_df_preprocessed = preprocessed(train_df)
```
