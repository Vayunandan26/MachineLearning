```python
import pandas as pd
import numpy as np
df = pd.read_csv('college_student_placement_dataset.csv')
df.head(5)
```

```python
df['CGPA'].mean()
```

```python
df['CGPA'] = df['CGPA'].mask(df['CGPA']>10, np.nan)
modified_mean = df['CGPA'].mean
df.fillna({'CGPA':modified_mean}, inplace= True)
```

```python
df['CGPA'].head(5)
df['CGPA'] = pd.to_numeric(df['CGPA'], errors='coerce')
```

```python
df['CGPA'].mean()
```

```python
df['Placement_Encoded'] = df['Placement'].map({'No': 0, 'Yes': 1})
```

```python
df['College_ID']= df['College_ID'].map(df.groupby('College_ID')['Placement_Encoded'].mean())
```
